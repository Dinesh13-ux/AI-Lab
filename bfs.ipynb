```python
from collections import deque
```

```python
def bfs(adj):
    V = len(adj)
    visited = [False] * V
    res = []
    
    src = 0
    q = deque()
    visited[src] = True
    q.append(src)

    while q:
        curr = q.popleft()
        res.append(curr)

        # visit all the unvisited
        # neighbours of current node
        for x in adj[curr]:
            if not visited[x]:
                visited[x] = True
                q.append(x)
                
    return res
```

```python
adj = [
    [1, 2],    
    [0, 3, 4], 
    [0, 5],    
    [1],       
    [1],       
    [2]        
]

print(bfs(adj))
```
