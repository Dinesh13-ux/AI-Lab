```python
from collections import deque
```

```python
def water_jug(a, b, target):
    visited = set()
    queue = deque([(0, 0, [])])  # (jugA, jugB, steps)

    while queue:
        x, y, steps = queue.popleft()

        if (x, y) in visited:
            continue
        visited.add((x, y))

        steps = steps + [(x, y)]

        # Goal check
        if x == target or y == target:
            return steps

        # Possible moves
        moves = [
            (a, y),                    # Fill Jug A
            (x, b),                    # Fill Jug B
            (0, y),                    # Empty Jug A
            (x, 0),                    # Empty Jug B
            (x - min(x, b - y), y + min(x, b - y)),  # A → B
            (x + min(y, a - x), y - min(y, a - x))   # B → A
        ]

        for move in moves:
            if move not in visited:
                queue.append((move[0], move[1], steps))

    return None
```

```python
a = int(input("Enter capacity of jug a: "))
b = int(input("Enter capacity of jug b: "))
c = int(input("Enter target capacity: "))
print("(a, b)")
result = water_jug(a, b, c)

if result:
    for step in result:
        print(step)
else:
    print("No solution")
```
